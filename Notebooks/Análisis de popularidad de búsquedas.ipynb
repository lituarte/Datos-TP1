## Análisis de popularidad de búsquedas

```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# Cargo las columnas que necesito
datos = pd.read_csv('/home/lautaro/Desktop/events.csv',
                    usecols = lambda col: col in ["event", "timestamp", "search_term"])
```

```python
busquedas = datos.loc[datos["event"] == "searched products",["timestamp","search_term"]]
busquedas = busquedas.loc[busquedas["search_term"].notnull(),:]
```

```python
text = ' '.join(busquedas["search_term"])

# Create and generate a word cloud image:
wordcloud = WordCloud(width=800, height=400, collocations=False).generate(text)

# Display the generated image:
plt.figure(figsize=[15,15])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
